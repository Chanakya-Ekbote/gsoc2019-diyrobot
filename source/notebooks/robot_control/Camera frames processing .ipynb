# Camera frames processing 

```python
from proteas_lib import vision
```

```python
canvas = vision.show_image(jupyter=True)
```

## Face detection

```python
face = vision.face_detection()
```

```python
cam_1 = vision.camera(camera=0)
for i in range(1,5):
    frame = cam_1.take_frame()
    final_frame = face.detect_face(frame)
    canvas.preview(final_frame)
cam_1.stop()
```

```python
print(face.get_pos())
```

```python
print(face.get_rect())
```

```python
print(face.get_faces()) # Return the number of the on the image
```
